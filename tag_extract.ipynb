```python
def extractNounChunks(doc):
    '''
    This function will extracts ners that are present with noun_chunks and also extracts all noun chunks with its root word from doc 
    •   Parameters : (doc)
    1.  doc   =  spacy doc object
    •   Returns :  ner_noun_chunk_list  = list  , noun_chunk_head_word  = nested list
    '''
    # Thi section to extract Ner_noun_chunks with ners
    ner_noun_chunk_list = []
    noun_chunks_lst = []
    
    for noun_chunk in doc.noun_chunks:
        noun_chunks_lst.append(noun_chunk)
    doc_ents_list = list(doc.ents)
    
    # To collect ner_span from doc.ents
    for ner_span in doc_ents_list:
        ner_span_index = range(ner_span.start,ner_span.end)
        # To collect noun_chunk_span from noun_chunk_list
        for noun_chunk_span in noun_chunks_lst:
            noun_chunk_index = range(noun_chunk_span.start,noun_chunk_span.end)
            # if ner_span intersects with noun_chunk_span, Collecting those noun_chunk and ner in list and appending to ner_noun_chunks list
            if(set(ner_span_index).intersection(noun_chunk_index)):
                ner_nc_lst = [str(noun_chunk_span), str(ner_span)]
                ner_noun_chunk_list.append(ner_nc_lst)
    
    # Thi section to extract All noun_chunks with their root words
    noun_chunk_head_word_list = []
    for noun_chunk in doc.noun_chunks:
        noun_chunk_head_word_list.append([noun_chunk.text, noun_chunk.root.text])
    return noun_chunk_head_word_list,ner_noun_chunk_list
```

```python
def mergeNers(docs_1,docs_2,doc2ieConfig,call_mode = None):
    """This function will merge ners present in custom model and default model if they are intersect or consecutively occurred mostly we give high priority to the custom model
    •   Parameters : (nlp_default_model, nlp_custom_model, text)
    1.  nlp_default_model  = spacy model
    2.  nlp_custom_model  =  custom spacy model
    3.  text   = str
    •   Returns :  default_model_docs,  noun_chunk_head_word_list, ner_noun_chunk_list, gv_ner_noun_chunk_list
    """
    
    if call_mode != 'LEGAL':
        ents_for_priority = ['ORG','FAC','LAW']
    else:
        ents_for_priority = ['COURT','PETITIONER','RESPONDENT','JUDGE','LAWYER','STATUTE','PROVISION','PRECEDENT','CASE_NUMBER','WITNESS']

    #ents_for_priority = doc2ieConfig['ents_for_priority'][call_mode]
    
    if call_mode == 'ESG':
        #Adding Custom Functions to manipulate ners
        docs_2 = VerifyOrgNers(docs_1,docs_2)
        docs_2 = RemoveMisdirectedNers(docs_2, ner_manipulation_dict, 'GV')
        docs_1 = RemoveMisdirectedNers(docs_1, ner_manipulation_dict, 'ORG')
    for ent in docs_2.ents:
        docs_1.vocab.strings.add(ent.label_)
    for ent in docs_1.ents:
        docs_2.vocab.strings.add(ent.label_)
    docs_1_ents_list = list(docs_1.ents)
    docs_2_ents_list = list(docs_2.ents)
    #Merged spans list from deault and custom models
    final_ents_list = docs_1_ents_list + docs_2_ents_list
    for docs_1_span in docs_1_ents_list:
        docs_1_span_index = range(docs_1_span.start,docs_1_span.end)
        for docs_2_span in docs_2_ents_list:
                docs_2_span_index = range(docs_2_span.start,docs_2_span.end)
                #Check for overlapping spans and giving priority to all custom spans unless default span is of ORG or FAC. ORG added through Entity Ruler(custom span) will have highest priority.
                if(set(docs_1_span_index).intersection(docs_2_span_index)):
                    # if custom_span.label_ == 'ORG' or default_span.label_ not in ['ORG','FAC']:
                    if (docs_2_span.ent_id) or (docs_1_span.label_ not in ents_for_priority):
                        if docs_1_span in final_ents_list:
                            final_ents_list.remove(docs_1_span)
                    else:
                        #Remove custom overlapping spans from the list
                        if docs_2_span in final_ents_list:
                            final_ents_list.remove(docs_2_span)
    #Assigning final ents list to default model
    if call_mode=="ESG":
        docs_1.ents = final_ents_list
        gv_ner_noun_chunk_list = extractGVnounchunks(docs_1)
        gv_ner_noun_chunk_list, gv_ner_noun_chunk_str_list = validatingNounChunks(gv_ner_noun_chunk_list)
        docs_1 = replacingGVwithNounChunks(gv_ner_noun_chunk_list, docs_1, ner_manipulation_dict)
        return docs_1, gv_ner_noun_chunk_str_list
    else:
        docs_2.ents = final_ents_list
        return docs_2
```

```python
def VerifyOrgNers(default_model_docs,custom_model_docs):
    '''
    Func description: Removes ORG ners if they dont start with UpperCase char or if its pos_tag is not NOUN or PROPER NOUN.
    Input params: default_model_docs - spacy doc object (created from default nlp_trf model)| custom_model_docs - spacy doc object (created from custom nlp model)
    Return values: custom_model_docs - spacy doc object (updated the ORG ners)
        '''

    #Removing "'s" from the Organizations when detected by model.
    ents_to_add = []
    ents_to_remove = []
    for ent in default_model_docs.ents:
        if ent.label_ == 'ORG' and ent[-1].text == "'s":
            new_ent = Span(default_model_docs, ent.start, ent.end-1, ent.label_)
            ents_to_remove.append(ent)
            ents_to_add.append(new_ent)

        if ent.label_ == 'ORG' and ent[0].text.lower() in ['the','a','an']:
            new_ent = Span(default_model_docs, ent.start+1, ent.end, ent.label_)
            ents_to_remove.append(ent)
            ents_to_add.append(new_ent)

    new_ents = [ent for ent in list(default_model_docs.ents) if ent not in ents_to_remove]
    default_model_docs.ents = set(filter_spans(new_ents + ents_to_add))

    # Checking if an org entity is properly detcted based on pos tag.   
    custom_ents_list = list(custom_model_docs.ents)
    for custom_ent in custom_model_docs.ents:
        if custom_ent.label_ == 'ORG':
            if str(custom_ent)[0].isupper(): 
                for token in default_model_docs:
                    if token.idx == custom_ent.start_char:
                        if token.pos_ in ['NOUN','PROPN','ADJ']:
                            pass
                        else:
                            custom_ents_list.remove(custom_ent)
                            break
            else:
                custom_ents_list.remove(custom_ent)
                
    #Assigning final ents list to custom model
    custom_model_docs.ents =  custom_ents_list
    
    return custom_model_docs
```

```python

def RemoveMisdirectedNers(docs, ner_manipulation_dict, ent_type):
    '''
    Func description: Removes Misdirected Organization ners by comparing the ner_manipulation_df - 1 column is used to remove exact match and another column is used to remove ners ending with misdirected terms.
    Input params: ner_manipulation_df - contains two required columns(Misdirected_ORG_Chunks and Misdirected_ORG) that contains ners to be removed from doc object |default_model_docs - spacy doc object (created from default nlp_trf model)| custom_model_docs - spacy doc object (created from custom nlp model)
    Return values: default_model_docs - updated spacy doc object after removing misdirected ORGs
    '''
    ents_list = list(docs.ents)
    col_name = 'Misdirected_'+ent_type
    chunk_col_name = 'Misdirected_'+ent_type+'_Chunks'

    for ent in docs.ents:
        if ent.text.lower().endswith(tuple(ner_manipulation_dict[chunk_col_name])) or ent.text.lower() in ner_manipulation_dict[col_name]:
            ents_list.remove(ent)
    docs.ents =  ents_list
    return docs
```

```python
def extractGVnounchunks(doc):
    '''
    Func description: Extracts ners that are present with noun_chunks and also extracts all noun chunks with its root word from doc
    Input params: doc - spacy doc object (final doc object)
    Return values: gv_ner_noun_chunk_list - nested list format(contains gv ner noun chunks with its respective gv ners - [gv_ner_with_noun_chunk, gv_ner])
        '''
    # Thi section to extract Ner_noun_chunks with ners
    gv_ner_noun_chunk_list = []
    noun_chunks_lst = []
    
    for noun_chunk in doc.noun_chunks:
        noun_chunks_lst.append(noun_chunk)
    doc_ents_list = list(doc.ents)
    
    # To collect ner_span from doc.ents
    for ner_span in doc_ents_list:
        if ner_span.label_ == 'GREEN_VARIABLE':
            ner_span_index = range(ner_span.start,ner_span.end)
            # To collect noun_chunk_span from noun_chunk_list
            for noun_chunk_span in noun_chunks_lst:
                noun_chunk_index = range(noun_chunk_span.start,noun_chunk_span.end)
                # if ner_span intersects with noun_chunk_span, Collecting those noun_chunk and ner in list and appending to ner_noun_chunks list
                if(set(ner_span_index).intersection(noun_chunk_index)):
                    ner_nc_lst = [noun_chunk_span, ner_span]
                    gv_ner_noun_chunk_list.append(ner_nc_lst)
    return gv_ner_noun_chunk_list
```

```python
def replacingGVwithNounChunks(gv_ner_noun_chunk_list, doc, ner_manipulation_dict):
    """
    Replaces GV NERs with their respective NounChunks based on certain constraints.

    1. The GV NounChunks must end with the GV NER (root word logic) and not occur in between them.
    2. Stopwords at the beginning of the chunks are avoided.
    3. If Measure related Ners are present in NounChunks, then those NounChunks are avoided.

    Args:
        gv_ner_noun_chunk_list (list): Nested list containing GV NERs and their respective NounChunks.
        doc (Doc): Spacy Doc object (final Doc object).
        ner_manipulation_dict (dict): Contains the required column "STOPWORDS".

    Returns:
        doc (Doc): Updated Spacy Doc object after replacing GV NERs with their NounChunks.
    """
    gv_noun_chunks = [nc[0] for nc in gv_ner_noun_chunk_list]
    measure_lst = [ent for ent in doc.ents if ent.label_ in ['CARDINAL', 'MONEY','QUANTITY', "PERCENT"]]
    def_org_lst = [ent for ent in doc.ents if ent.label_ in ['NORP', 'ORG','FAC']]
    ents_to_remove = []
    ents_to_add = []

    for ent in doc.ents:
        if ent.label_ == "GREEN_VARIABLE":
            for noun_chunk in gv_noun_chunks:
                count = 0
                for measure in measure_lst:
                    if measure.start in range(noun_chunk.start,noun_chunk.end) or measure.end in range(noun_chunk.start,noun_chunk.end):
                        count += 1
                
                for org in def_org_lst:
                    if org.start in range(noun_chunk.start,noun_chunk.end) or org.end in range(noun_chunk.start,noun_chunk.end):
                        count += 1
                if count > 0:
                    continue
                if ent.end == noun_chunk.end and noun_chunk.start < ent.start:
                    nc_token_index = 0
                    if str(noun_chunk[0]) in ['%','$']:
                        nc_token_index += 1
                    for token in noun_chunk:  
                        
                        if token.text.lower() in ner_manipulation_dict['STOPWORDS']:
                            nc_token_index += 1
                        else:
                            break
                    new_span_start = noun_chunk.start + nc_token_index
                    new_ent_span = Span(doc, new_span_start, noun_chunk.end, ent.label_)
                    ents_to_remove.append(ent)
                    ents_to_add.append(new_ent_span)
    new_ents = [ent for ent in list(doc.ents) if ent not in ents_to_remove]
    doc.ents = set(filter_spans(new_ents + ents_to_add))
    return doc
```

```python
def validatingNounChunks(gv_ner_noun_chunk_list):
    '''
    Func description: Validate the GV_ner_nounchunks nested list and modifying spans or removing based on some constraints.
        The constraints used are : GV noun chunk must be ending with the GV ner, The tokens in noun chunks must be either NOUN or PROPN or ADJ, Both Noun and Adj should not occur more than once togethor.
    Input params: gv_ner_noun_chunk_list - nested list with all [GV_ner_noun_chunk, GV_ner] as elements.
    Return values: gv_ner_noun_chunk_list - nested list with valid [GV_ner_noun_chunk, GV_ner] as elements.
        '''
    valid_gv_nc_lst = []
    valid_gv_nc_str_lst =[]
    for nc_list in gv_ner_noun_chunk_list:
        # Get the length of the root word
        len_of_root_word = len(nc_list[1])
        # Initialize the variables to store the count of adjectives and nouns
        adj_count = 0
        noun_count = 0
        rev_token = []
        if str(nc_list[0]).endswith(str(nc_list[1])):
            # Loop through the reversed list of tokens
            for index, token in enumerate(reversed(nc_list[0])):
                if index > len_of_root_word-1 and token.pos_ in ['NOUN','PROPN','ADJ']:
                    # Increment the adjective and noun count
                    adj_count += (token.pos_ == 'ADJ')
                    noun_count += (token.pos_ == 'NOUN')
                    # If there are more than 1 adjective and 1 noun or more than 1 noun and 1 adjective, break
                    if adj_count > 1 and noun_count >= 1 or noun_count > 1 and adj_count >= 1:
                        break
                    rev_token.append(token)
                # If the token is not a noun, adjective or proper noun, break
                elif index > len_of_root_word-1:
                    break
            # If there are tokens stored in rev_token
            if rev_token:        
                span_start = [t for t in nc_list[0] if t == rev_token[-1]][0]
                index = span_start.i - nc_list[0].start
                span = nc_list[0][index:nc_list[1].end]
                valid_gv_nc_lst.append([span, nc_list[1]])
                valid_gv_nc_str_lst.append([str(span),str(nc_list[1])])
            else:
                span = nc_list[1]
                valid_gv_nc_lst.append([span, nc_list[1]])
                valid_gv_nc_str_lst.append([str(span),str(nc_list[1])])
    return valid_gv_nc_lst, valid_gv_nc_str_lst
    
```

```python
def text_preprocessing(text, use_case = None):
    try:
        '''Text preprocessing steps:
            1. Removing html links
            2. removing html tags
            3. Remove accented characters from text, For example : café to cafe.
            4. Removing Emoji and other pictographs
            5. Filtering string and getting only alphabets,numbers,comma,fullstop,-.
            6. Joining lines based on line breaks with full stops'.'.
            7. Removing whitespace
            8. Adding a space before the starting of a new sentence
            9. matching for fullstops that dont have space after them and correcting them
            10. Looking for abbrevations with space after fullstops and correcting them'''
        #The text preprocessing function has been updated to add fullstops when there is a line break in documents. It has improved ner noun chunk detection. Also we are handling exceptions on adding space after fullstops.
        string = ""
        text = removeAccentedChars(text)
        text = removeEmoji(text)
        #removing exceptions for the above code using replace option. Please add more exceptions when you find it.
        # exception_str = {"https:// ":"https://",'. com':'.com',' .com':'.com','www. ':'www.','www .':'www.'}
        url_exception_str = doc2ieConfig['exceptions_dict']['url_exception_str']
        for key,value in url_exception_str.items():
            text = text.replace(key,value)
        remove_lead_text = deepcopy(text)
        text = remove_lead_text
        #Removing email IDs
        text = re.sub(r"(\S*@\S*\s?)" ,"",text)
            # remove html markup tags
        text= re.sub("(<.*?>)","",text)
    #     text =re.sub('http://\S+|https://\S+|www\S+', ' ', text)
        text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»""'']))''', " ", text)
        # filtering string and getting only alphabets,numbers,comma,fullstop,-.
        text = re.sub("[^₹¥€£$%'=)(,&/a-zA-Z0-9\.?:-]/n", ' ', text)
        #if use case is Doc then adding fullstop is adding at the end of each line
        if use_case == "Doc" or use_case == "txt" or use_case == "pdf" or use_case == 'csv':
            text_new = ''
            for line in text.splitlines():
                line = line.strip()
                if line != '':
                    if line[-1] not  in ['.',':']:#if line[-1]!= '.':
                        line+="."
                    text_new= text_new + line + " "
            text = text_new
            text = text.strip()
            text = re.sub('[.]+', '.',text_new)     # Replacing one or more consecutive fullstops with a single fullstop.

        #remove whitespace
        text=text.strip()
        if not str(text).endswith(".jpg"):
            text = re.sub(r'--+(?!-)', ' ', text) # removed multiple hyphens and keeping single hyphen
        text = text.replace(" .",".")
        # adding a space before the starting of a new sentence.
        pattern = re.findall('[a-z]\.[A-Z]|[0-9]\.[A-Z]|[a-z]\.[a-z]',text)
        for str2 in pattern:
            str3 = str2.replace(".",". ")
            text = text.replace(str2,str3)

        # exception_str = {"₹":"INR","¥":"JPY","£":"GBP","$":"USD","€":"EUR","USDUSD":"USD "}
        money_exception_str = doc2ieConfig['exceptions_dict']['money_exception_str']
        for key,value in money_exception_str.items():
            text = text.replace(key,value)

        # sentence tokenize on text
        sents = sent_tokenize(text)
        for sent in sents:
            # matching for fullstops that dont have space after them and correcting them.
            for word in sent.split():
                word = seperateNumChars(word)
                wrong_match = re.findall(r'(?:[A-Z]\.)+',word) # checking for abbreviations
                match = re.findall("\.[A-Z]",word)  # matching with
                if len(match) and not len(wrong_match):
                    idx  = word.index(match[0])
                    word = word.replace(word[idx],". ")
                    word = word.strip()
                    string = string +" "+ word
                    string = string.strip()
                else:
                    string = string +" "+ word
                    string = string.strip()
        text = string
        
        
        #     # #         #removing exceptions for the above code using replace option. Please add more exceptions when you find it.
        # exception_str = {"N0. 1":"No.1","Gov. ":"Gov.","No. 1":"No.1","L.L.C. ":"L.L.C.","B.C. ":"B.C.","FSMA.":"FSMA. ","Pte. Ltd. ":"Pte.Ltd",'. com':'.com',' .com':'.com','. org':'.org',' .org':'.org','. co':'.co',' .co':'.co','Mr. ':'Mr.','Mr .':'Mr.','Mrs. ':'Mrs.','Mrs .':'Mrs.','Dr. ':'Dr.','Dr .':'Dr.','Prof. ':'Prof.','Hon. ':'Hon.', 'Prof .':'Prof.','. uk':'.uk','the.financial':'the financial','the. financial': 'the financial'}
        general_exception_terms = doc2ieConfig['exceptions_dict']['general_exception_terms']
        for key,value in general_exception_terms.items():
            text = text.replace(key,value)

        remove_lead_text = deepcopy(text)
        counter = 0
        for idx, char in enumerate(text):
            if char in ['-','.',' ']:
                remove_lead_text = remove_lead_text[:idx+counter] + remove_lead_text[idx+1+counter:]
                counter-=1
            else:
                break
        text = remove_lead_text
        text = add_space_before_punctuation(text)
        return {"status":"sucess","response":text}
    except Exception as err:
        print("error: ",traceback.print_exc())
        return {"status":"error","message":"Server busy please try after some time."}
```

```python
from unidecode import unidecode
def removeAccentedChars(text):
    # Define a translation table to replace all characters that are not money symbols with their ASCII equivalents
    trans_table = str.maketrans({char: unidecode(char) if not (char in '$€₹£¥') else char for char in text})
    # Translate the input string using the translation table
    text = text.translate(trans_table)
    text = text.replace('US$','USD')
    return text
def removeEmoji(text):
    '''
    Func description: Removes emojis from text
    Input params: text - str format
    Return values: text - str format (Updated text)
        '''
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    return text
```

```python
import spacy
from spacy.tokens import Span
from spacy.util import filter_spans
from spacy.pipeline import merge_entities
from spacy.language import Language
import pandas as pd
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import warnings
warnings.filterwarnings("ignore")

# Local module imports 
import config_manager
#import preprocessing_text as tp


__nlp_default_model = None
__nlp_custom_model = None
__nlp_esg_bert = None
__esg_sent_model = None
__nlp_legal_model = None
__nlp_crime_model = None

#doc2ieConfig = config_manager.get_configs(0)[1]
f = open('hd_doc2ie_config_esg.json')
doc2ieConfig = json.load(f)

def __initiliazeModels():
    # cfgObj = config_manager.configManager()
    # ontoConfig = cfgObj.getConfig("ontoConfig")
    print("initiliaze Models..")
    global __nlp_default_model
    global __nlp_custom_model
    global __nlp_esg_bert
    global __esg_sent_model
    global __nlp_legal_model
    global __models_obj
    global __nlp_crime_model

    tokenizer = AutoTokenizer.from_pretrained("G://ESG-BERT")
    model = AutoModelForSequenceClassification.from_pretrained("G://ESG-BERT")
    __nlp_esg_bert = pipeline("text-classification", model=model, tokenizer=tokenizer)
    sent_tokenizer = AutoTokenizer.from_pretrained("G://sent_tokenizer")

    sent_model = AutoModelForSequenceClassification.from_pretrained("G://sent_tokenizer")
    __esg_sent_model = pipeline("text-classification", model=sent_model, tokenizer=sent_tokenizer)

    '''Loading Flan-t5-model for question answer and summarization'''
    # if(ontoConfig["llm"]["model_type"] == "flan-t5"):
    # if False:
    #     from transformers import T5Tokenizer, T5ForConditionalGeneration
    #     flank_modelName = ontoConfig["llm"]["flank_modelName"]
    #     print("Loading flan-t5 model....",flank_modelName)
    #     flan_t5_model_path = "../custom_models/flan-t5-models/"+flank_modelName
    #     __QA_tokenizer = T5Tokenizer.from_pretrained(flan_t5_model_path)
    #     __QA_model = T5ForConditionalGeneration.from_pretrained(flan_t5_model_path)

    custom_ners_filepath = 'curated_custom_ners.csv'
    '''Loading default nlp modes'''
    print("Default spacy trf_model custom model....")
    __nlp_default_model = spacy.load("en_core_web_trf")

    if "mergeEntities" not in __nlp_default_model.pipe_names:
        '''Adding custom component to merge ners of same label'''
        __nlp_default_model.add_pipe("mergeEntities", last = True)

    print("Loading custom model....")
    custom_model_path= "G://climate_finance_ner//model-best"
    __nlp_custom_model = spacy.load(custom_model_path)

    legal_nlp_model_path= "G://legal_nlp"
    __nlp_legal_model = spacy.load(legal_nlp_model_path)
    
    crime_nlp_model_path = "G://crime_ner_model//model-best"
    __nlp_crime_model = spacy.load(crime_nlp_model_path)
        
    #If Update is needed, Call this function Or you can comment it out.
    __nlp_custom_model = UpdateModelwithERuler(__nlp_custom_model, custom_ners_filepath)
    
    if "remove_specific_custom_ners" not in __nlp_custom_model.pipe_names:
        '''Adding custom component to remove exceptions from doc.ents'''
        __nlp_custom_model.add_pipe("remove_specific_custom_ners", last = True)
        
    if "removeAddException" not in __nlp_custom_model.pipe_names:
        '''Adding custom component to remove exceptions from doc.ents'''
        __nlp_custom_model.add_pipe("removeAddException", last = True)
    
    if "PostprocessingOrgNers" not in __nlp_custom_model.pipe_names:
        '''Adding custom component to merge ners of same label'''
        __nlp_custom_model.add_pipe("PostprocessingOrgNers", last = True)
        
        
    if "mergeEntities" not in __nlp_custom_model.pipe_names:
        '''Adding custom component to merge ners of same label'''
        __nlp_custom_model.add_pipe("mergeEntities", last = True)
     
    
        
        
    #Loading Qg model
    # print("Loading QG model....")
    # __QG_tokenizer = AutoTokenizer.from_pretrained("mrm8488/t5-base-finetuned-question-generation-ap")
    # __QG_model = AutoModelWithLMHead.from_pretrained("mrm8488/t5-base-finetuned-question-generation-ap")
    
    print("default: ",__nlp_default_model.pipe_names)
    print("Custom: ",__nlp_custom_model.pipe_names)
    print("Crime: ",__nlp_crime_model.pipe_names)
    
    __models_obj = {"spacy_default_model":__nlp_default_model,"esg_custom_model":__nlp_custom_model,"spacy_legal_model":__nlp_legal_model, "spacy_crime_model": __nlp_crime_model, "esg_sent_model":__esg_sent_model,"esg_bert":__nlp_esg_bert}
    
    

#Updating Model Function
def UpdateModelwithERuler(__nlp_custom_model, custom_ners_filepath):
    ners_to_detect = ReadCustomNers(custom_ners_filepath)
    __nlp_custom_model = addCustomNerEntities(__nlp_custom_model, ners_to_detect)
    return __nlp_custom_model
    
    
    
#Add ners to the entity ruler using patterns (matcher)
def addCustomNerEntities(nlp, ners_to_detect):
    '''
    Func description: Create Entity_ruler component to the nlp object, then add ners and ner_labels as ruler patterns.
    Input params: nlp - spacy nlp object| ners_to_detect - list of dict containing ner & ner_label as key:value pairs.
    Return values: nlp - updated spacy nlp object with entity ruler component'''
    #If entity ruler not in pipeline, then add it or else skip this function.
    count = 0
    if("entity_ruler" not in nlp.pipe_names): 
        ruler = nlp.add_pipe("entity_ruler",config = {"overwrite_ents":True}) # Overwriting model detected ners
    else: 
        nlp.remove_pipe("entity_ruler")
        ruler = nlp.add_pipe("entity_ruler",config = {"overwrite_ents":True}) # Addding fresh E ruler component
    # trailing_tokens = ['corporation','ltd.','corp.','inc.','pvt','pvt.','limited','corp', 'co','inc','group', 'ltd',"llc","private","project","plant"]
    # trailing_punc = ['-','.']
    trailing_tokens = doc2ieConfig['trailing_terms']['trailing_tokens']
    trailing_punc = doc2ieConfig['trailing_terms']['trailing_punc']
    for item in tqdm(ners_to_detect):
        # key - ner and item[key] - ner_label
        key = list(item.keys())[0]
        processed_key = text_preprocessing(key)['response']
        items_list = nlp.make_doc(processed_key)
        exact_string_with_excep = [{"LOWER" : str(token).lower()} if not str(token) in ['-','.'] else {'ORTH':str(token)} for token in items_list]
        # Looking for trailing tokens if the label is ORG
        if str(item[key]) == 'ORG': 
            
            exact_string_with_excep.extend([{"LOWER":{"IN": trailing_tokens}, "OP":"?"}])
            exact_string_with_excep.extend([{"ORTH": {"IN": trailing_punc}, 'OP':'?'}])
            exact_string_with_excep.extend([{"LOWER" : {"IN":trailing_tokens}, "OP":'?'}])
        # patterns = [{"label": str(item[key]), "pattern" : exact_string_with_excep}]
        patterns = [{"label": str(item[key]), "pattern" : exact_string_with_excep, "id":count}]
        ruler.add_patterns(patterns)
        count += 1

    return nlp
    
    
    
    #Reading the file from ner csv file:
def ReadCustomNers(file_path):
    '''
    Func description: Reads the ners csv file and converts to the required format to be added to the entity ruler
    Input params: file_path - .csv or excel format| path of the custom ners csv file (contains ner, ner_type and ner_mapping)
    Return values: csv_list - list format| returns list of dictionaries(each ner & ner_type as key:value pairs)
        '''
    list_of_dict=[]
    if  file_path.endswith('.csv'):
        df=pd.read_csv(file_path)
    elif file_path.endswith('.xlsx'):
        df=pd.read_excel(file_path,sheet_name = 'ner and ner_type') # Used for .excel format
    read_data=df.values.tolist()
    for i in read_data:
        list_of_dict.append({i[0]:i[1]})
    return list_of_dict




#Remove Investment and Collaboration entities
@Language.component('remove_specific_custom_ners')
def remove_specific_custom_ners(doc):
    remove_specific_labels = doc2ieConfig['remove_specific_labels']
    spans = []
    for ent in doc.ents:
        if ent.label_ not in remove_specific_labels:
            spans.append(ent)
    doc.ents = spans
    return doc
    
    
@Language.component('PostprocessingOrgNers')
def PostprocessingOrgNers(doc):
    '''
    Func description: Custom component to post process any exception regarding ORG ners.
                      Here we are removing punctuation at the end of the ORG ners if they dont belong to any trailing exception.
    Input params: doc - spacy doc object
    Return values: doc - spacy doc object (Updated)
        '''
    # trailing_excep_token_lst = ['co.','ltd.','corp.','inc.','pvt','pvt.','corp', 'co','inc', 'ltd',"llc"]
    # trailing_excep_punc_lst = [".","-",","]
    trailing_excep_token_lst = doc2ieConfig['trailing_excep_terms']['trailing_excep_token_lst']
    trailing_excep_punc_lst = doc2ieConfig['trailing_excep_terms']['trailing_excep_punc_lst']
    ents_to_remove = []
    entity_lst =[]
    add_ents = []

    for ent in doc.ents:
        if (ent.label_ == 'ORG') and any(char in trailing_excep_punc_lst for char in str(ent)):
                entity_lst.append(ent) #entity list contains ners with punc

    for ent in entity_lst: 
        if str(ent) in trailing_excep_punc_lst: #if ent wrongly matched as punct, remove them
            ents_to_remove.append(ent)
            continue

        split_lst = [ent.text[0:-1]] # assuimg the last token to be the punc
        root_word_ner=''
        # We extract only the last token before the punc and check it with our trailing exceptions tokens, if they are not in them then update the ORG ner.
        for ele in split_lst:
            for char in ele:
                root_word_ner+=char
                if char == ' ':
                    root_word_ner = ''
            if root_word_ner.lower() in trailing_excep_token_lst:
                pass
            else:
                Remove_span = Span(doc, ent.start, ent.end, ent.label_)
                Add_span = Span(doc, ent.start, ent.end-1, ent.label_)
                add_ents.append(Add_span)
                ents_to_remove.append(Remove_span)
 
    total_lst = add_ents+ list(doc.ents) #adds the newer entries
    # new_total_lst = [ele for ele in total_lst if ele not in ents_to_remove] #removing the older entities.
    new_total_lst = [ele for ele in total_lst if ele.text not in [e.text for e in ents_to_remove]] #removing the older entities.
    filtered_ents = tuple(filter_spans(new_total_lst)) #filter duplicates
    doc.ents = filtered_ents
    return doc
    
    
@Language.component('removeAddException')
def removeExceptionComponent(doc):
    ents_to_remove = []
    # remove_ents_list = ['Elect',  'Sess',  'Easter',  'Jet',  'railway lines',  'Wulf',  'jet',  'Wildlife',  'Ale',  'defence',  'land',  'Gross',  'goss',  'railway station',  'tune',  'Signatures',  'Transformation',  'Tweet',  'jewellery',  'Fabric curators',  'green bond',  'factory',  'Preservation',  'Freedom',  'Animals',  'factories',  'Outdoors',  'fibre 1.5',  'consumption',  'Oat',  'fibre can',  'FURNITURE',  'transformation',  'Emile',  'lands',  'Pink',  'Tear',  'Clean',  'DOCUMENTS',  'Timing',  'henkel',  'renew',  'Land',  'Eligibility',  'BEAT',  'conservation',  'petri',  'clean',  'green',  'processors',  'Surprise',  'Facemasks']
    remove_ents_list = doc2ieConfig['remove_ents_list']
    for ent in doc.ents:
        if ent.text in remove_ents_list:
            ents_to_remove.append(ent)
            
    total_ent_lst = list(doc.ents)
    new_total_ent_lst = [ele for ele in total_ent_lst if ele not in ents_to_remove] #removing the older entities.
    filtered_ents = set(filter_spans(new_total_ent_lst)) #filter duplicates
    doc.ents = filtered_ents
    return doc
    
    
@Language.component("mergeEntities")
def mergeEntities(doc):
    '''
    Func description: Custom made merging entities function to merge two ners of same label occuring togethor.
    Input params: doc - spacy doc object
    Return values: doc - spacy doc object (Updated spacy doc object after merging entities)
        '''
    
    #initial default values for token for first iteration
    temp_end = 0
    temp_label_ = ''
    temp_text = ''
    temp =''
    new_ents = []
    ents_to_remove = []

    for ent in doc.ents:
        if not ent.label_ in ['ORG','PERSON']:
            if((temp_end == ent.start) and (temp_label_ == ent.label_)) or ((temp_end == ent.start-1) and (temp_label_ == ent.label_) and (str(doc[temp_end:ent.start]) in ['-','and'])): 
                #if identified two words of same entities next to next, then we add the span of both entities in add_span to be added to doc.ents and we add the span of those individual entities which has to be removed.
                add_span = Span(doc, temp_start, ent.end, ent.label_) #combined entity span
                new_ents.append(add_span)
                rem_span1 = Span(doc, temp_start, temp_end, ent.label_) #previous entity span
                rem_span2 = Span(doc, ent.start, ent.end, ent.label_)  #current entity span
                ents_to_remove.append(rem_span1)
                ents_to_remove.append(rem_span2)

                temp_start = temp_start  #reassigning the current entity data to the temporary variables.
                temp_end = ent.end
            else:
                temp_start = ent.start
                temp_end = ent.end
            temp = ent
            temp_text = ent.text
            temp_label_ = ent.label_
        else:
            if((temp_end == ent.start) and (temp_label_ == ent.label_) and str(doc[ent.start:ent.end])!=str(temp_text)): 
                #if identified two words of same entities next to next, then we add the span of both entities in add_span to be added to doc.ents and we add the span of those individual entities which has to be removed.
                add_span = Span(doc, temp_start, ent.end, ent.label_) #combined entity span
                new_ents.append(add_span)
                rem_span1 = Span(doc, temp_start, temp_end, ent.label_) #previous entity span
                rem_span2 = Span(doc, ent.start, ent.end, ent.label_)  #current entity span
                ents_to_remove.append(rem_span1)
                ents_to_remove.append(rem_span2)

                temp_start = temp_start  #reassigning the current entity data to the temporary variables.
                temp_end = ent.end
            else:
                temp_start = ent.start
                temp_end = ent.end
            temp = ent
            temp_text = ent.text
            temp_label_ = ent.label_

    total_lst = new_ents+ list(doc.ents)
    new_total_lst = [ele for ele in total_lst if ele not in ents_to_remove] #removing the older entities.
    filtered_ents = tuple(filter_spans(new_total_lst)) #filter duplicates
    doc.ents = filtered_ents
    return doc
```

```python
def CustomDocTokenizer(doc, tokenized_sents):
    '''
    Splitting the spacy Doc object using the custom sentence tokenizer used for str data type
    This function will preserve the entities and other attributes on the split sentences.
    Provide a recursive limit for how many potential delimiters you could encounter.
    If span becomes None, We will miss that particular sentence which can happen during exception (no such scenario found yet).

    input : spacy Doc object
    output : Updated spacy Doc object with splitting component
    '''
    count = 0
    span_lst = []
    recursive_limit = 5
    
    for sent in tokenized_sents:
        start = deepcopy(count)
        length = len(sent)  # Calculate the length of the sentence
        
        # Calculate the end index by adding the length of the sentence
        end = start + length

        # print(end)
        span = doc.char_span(start, end)
        
        if span is not None:
            span_lst.append(span)
        # else:
        #     if sent == '':
        #         print('empty text block skipped..')
        #     else:
        #         print('************Sentence missed because of span mismatch************')
        count = end + 1   
    doc.set_extension("sentences", default=span_lst, force=True)
    return doc
```

```python
def RecursiveFunction(sent_char,doc,count,limit,current=0):
    if current >= limit:
        raise RecursionError("Recursion limit reached | Potentially splitting function modified Text in sentence")
    if sent_char == doc.text[count] :
        return count
    else:
        count+=1
        current+=1
        return RecursiveFunction(sent_char,doc,count,limit,current)
```
